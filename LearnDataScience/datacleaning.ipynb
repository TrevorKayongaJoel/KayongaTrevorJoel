```python
# Libraries for Data manupilation

import os
import pandas as pd
import numpy as np


# Load a data set
df = pd.read_csv("data.csv")
```

```python
df.head(5)
```

```python
# Data Preprocessing or Data Cleaning
# Bad data - example
# empty cell - cell has no record
# wrong format 
# wrong data, 
# Duplicates
```

```python
# Empty cells
# Check for empty values
print(df.isnull().sum())
```

```python
# Check for wrong format
print(df.dtypes)
```

```python
# Check for non numeric values in the columns
for col in ["Duration", "Pulse", "Maxpulse","Calories"]:
    if not pd.to_numeric(df[col], errors="coerce").notnull().all():
        print(f"Non numeric data found in column: {col}")
```

```python
# Wrong data
# Check for negative values in numeric columns 
for col in ["Duration", "Pulse", "Maxpulse","Calories"]:
    if (df[col] < 0).any():
        print(f"Negative values found in column: {col}")
    else:
        print("No negative values found")    
        
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))

# Histogram of "Duration"
plt.subplot(2, 2, 1)
sns.histplot(df["Duration"], bins = 30, kde=True)
plt.title("Duration Distribution")
# Histogram of "Pulse"
plt.subplot(2, 2, 2)
sns.histplot(df["Pulse"], bins = 30, kde=True)
plt.title("Pulse Distribution")
```

```python
# Scatter plot of Duration vs Pulse
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.scatterplot(x="Duration", y = "Pulse", data = df)
plt.title("Duration Vs Pulse")
# Scatter plot of  Pulse vs caloris
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 2)
sns.scatterplot(x="Pulse", y = "Calories", data = df)
plt.title("Pulse vs Calories")
```
